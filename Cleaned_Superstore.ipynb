Check for Missing Values

```python
# Total missing values per column
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

Check & Convert Data Types

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

Remove Duplicates

```python
# Check for duplicates

df.duplicated().sum()
```

Outlier Detection

```python
# Basic descriptive stats
df[['Sales', 'Profit', 'Discount']].describe()

# You can visualize with boxplot
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data=df[['Sales', 'Profit', 'Discount']])
plt.show()
```

```python
# Remove outliers based on z-score
from scipy import stats
df = df[(np.abs(stats.zscore(df[['Sales', 'Profit']])) < 3).all(axis=1)]
```

Clean Categorical Columns

```python
# Standardize text format

df['City'] = df['City'].str.title()
df['State'] = df['State'].str.title()
df['Category'] = df['Category'].str.title()
```

Feature Engineering (New Columns)

```python
# Extract year and month from order date
df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day_name()
```

Final Check & Save Cleaned File

```python
# Final shape and info
df.info()

# Save cleaned file
df.to_csv('Cleaned_Superstore.csv', index=False)
```
